# Mental Health Risk Prediction Using Machine Learning

## Research Question

Can demographic, academic, and personal characteristics be used to predict whether a college student reports experiencing depression?

## Project Objective

The objective of this project is to explore the relationships between demographic and academic factors and student mental health. By analyzing survey data and applying machine learning techniques, this project aims to identify patterns associated with depression among college students and determine which factors contribute most to predicting mental health outcomes.

## Hypotheses

### Hypothesis 1
Students with different CGPA ranges may have different rates of depression.

**Reason:** Academic performance may be associated with stress and overall mental well-being.

### Hypothesis 2
Depression rates may vary across years of study.

**Reason:** Academic workload, responsibilities, and life transitions change as students progress through college.

### Hypothesis 3
Age and gender may be associated with differences in reported depression.

**Reason:** Mental health experiences can vary across demographic groups.

### Additional Variable I Would Include

If I could redesign this survey, I would add a question about average sleep duration.

**Reason:** Sleep habits are strongly associated with stress, anxiety, and depression and would likely improve the model's predictive ability.

"The classification report provides class-specific evaluation metrics, including precision, recall, F1-score, and support. In my project, the model achieved 71.43% overall accuracy.
